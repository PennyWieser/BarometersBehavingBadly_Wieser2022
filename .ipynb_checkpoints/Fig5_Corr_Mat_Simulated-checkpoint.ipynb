```python
# Drop columns for consistency with other correlation figures
Cpxs_Expsam2_d=Cpxs_Exp2.drop(columns=['K2O_Cpx', 'Sample_ID_Cpx'])
Cpxs_Expsam2_dc=Cpxs_Expsam2_d.copy()
Cpxs_Expsam2_dc.columns=Cpxs_Expsam2_d.columns.str.replace('_Cpx', '')

# Makes neat dataframe of Cpx stuff for correlation matrix
Cpxs_Expsam2_d_neat1=Cpxs_Expsam2_dc.rename(columns={'SiO2': 'SiO$_2$',
                                                      'TiO2': 'TiO$_2$',
                                                          'Al2O3': 'Al$_2$O$_3$',
                                                         'FeOt': 'FeO$_t$',
                                                         'Cr2O3': 'Cr$_2$O$_3$',
                                                         'Na2O': 'Na$_2$O'
                                                      })


```

## Use noise functions to create a synthetic noise based on standard deviation and mean

```python
# Calculate Std Dev of this experiment for every element
Cpxs_Std=Cpxs_Exp2.std(axis=0).to_frame().T

# Add err ready to pass into function. 
Cpxs_Std=Cpxs_Std.add_suffix('_Err')

# Mean of dataframe
Cpxs_Av=Cpxs_Exp2.mean(axis=0).to_frame().T

# Saying use absolute errors, add normally distributed noise to the mean. Make 1000 datapoints
Cpxs_Exp2_Synthetic=pt.add_noise_sample_1phase(phase_comp=Cpxs_Av, phase_err=Cpxs_Std,
 phase_err_type="Abs", duplicates=1000, err_dist="normal", positive=True)

Cpxs_Exp2_Calc_Synthetics=pt.calculate_cpx_only_press_all_eqs(cpx_comps=Cpxs_Exp2_Synthetic)
```

```python
import seaborn as sns
fig, ((ax10, ax1, ax2)) = plt.subplots(1,3, figsize = (15,5)) # adjust dimensions of figure here

# Make 
sns.heatmap(Cpxs_Expsam2_d_neat1.corr()**2,  vmax=1, square=True, 
            vmin=0, center= 0.5, cmap= 'rocket_r',
            annot=True, annot_kws={"size":10}, fmt='.2f', ax=ax10, cbar_kws={"shrink": .75});



ax1.plot(Cpxs_Exp2_Calc_Synthetics['TiO2_Cpx'], Cpxs_Exp2_Calc_Synthetics['Al2O3_Cpx'], '.y', mfc='yellow')
ax1.plot(Cpxs_Exp2['TiO2_Cpx'], Cpxs_Exp2['Al2O3_Cpx'], 'ok', mfc='red', ms=7)

ax2.plot(Cpxs_Exp2_Calc_Synthetics['Na2O_Cpx'], Cpxs_Exp2_Calc_Synthetics['MnO_Cpx'], '.y', mfc='yellow')
ax2.plot(Cpxs_Exp2['Na2O_Cpx'], Cpxs_Exp2['MnO_Cpx'], 'ok', mfc='red', ms=7)

ax1.set_xlabel('TiO$_2$ Cpx (wt%)')
ax1.set_ylabel('Al$_2$O$_3$ Cpx (wt%)')
ax2.set_xlabel('Na$_2$O Cpx (wt%)')
ax2.set_ylabel('MnO Cpx (wt%)')
fig.suptitle("Experiment Y0201-2, Neave et al. (2019), N=" + str(len(Cpxs_Expsam2_d_neat1)), fontsize=14)
fig.tight_layout()
fig.savefig('Neave_Corr_SiTi_NaMn.png', dpi=200)
```
