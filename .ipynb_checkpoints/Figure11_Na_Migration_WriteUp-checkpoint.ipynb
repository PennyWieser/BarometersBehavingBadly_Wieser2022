```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.max_columns = None
from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
def calculate_R2(x, y):

    masknan=(~np.isnan(x) & ~np.isnan(y))
    regx=x[masknan].values.reshape(-1, 1)
    regy=y[masknan].values.reshape(-1, 1)
    slope, intercept, r_value, p_value, std_err = stats.linregress(regx[:,0],regy[:,0])
    p_value=np.round(p_value, 3)
    lr=LinearRegression()
    lr.fit(regx,regy)
    Y_pred=lr.predict(regx)
    Int=lr.intercept_
    Grad=lr.coef_
    #R="R\N{SUPERSCRIPT TWO} = " +  str(np.round(r2_score(regy, Y_pred), 2))
    R=(np.round(r2_score(regy, Y_pred), 2))
    RMSE=std_dev(regx, regy)
    RMSEp=np.round(RMSE, 2)
    Median=np.nanmedian(regy-regx)#
    Medianp=np.round(Median, 2)
    Mean=np.nanmean(regy-regx)
    Meanp=np.round(Mean, 2)


    return {'R2': '{0:.2f}'.format(R), 'RMSE':'{0:.2f}'.format(RMSEp), 'RMSE_num':RMSEp,
    'P_val':'{0:.3f}'.format(p_value), 'Median':'{0:.2f}'.format(Medianp), 'Mean':'{0:.2f}'.format(Meanp),
'x_pred': regx, 'y_pred': Y_pred, 'Int': Int, 'Grad':Grad[0]}

def std_dev(x, x2):
    N=len(x)
    sum_squares=np.sum((x-x2)**2)
    RMSE=np.sqrt((1/N)*sum_squares)
    return RMSE
```

## Method 1 - Tracking declining intensities using Peak-Background subcounting routines
- We use a subcounting routine on the PeakFit Cameca software, where the Peak - bck(-) - bck (+) is measured 10 separate times during the routine. Although we ask the software to count for 60s on Na, using the log file, we can see by the time the spectrometer had switched from peak-bck-bck 10 times, the total time the beam was on while collecting on Na was ~260 s. 

### Tracking declining intensities in Rhyolite to validate our method.
- Here, we analyse a hydrous rhyolitic glass with ~3 wt% H$_2$O, as we expect to see beam migration in this material
- It is clear that Na is rapidly migrating at these beam conditions, and this subcounting routine is capturing this change

```python
Rhyo_test=pd.read_excel('Supporting_Data_10.xlsx', sheet_name='rhyo_test1_I_decay')
```

```python
fig, (ax1) = plt.subplots(1, 1, figsize=(7,5))
ax1.set_title('Testing Na migration in a hydrous rhyolite glass at 20 nA')
ax1.plot(Rhyo_test['time since beam turned on'].iloc[0:10], 
         Rhyo_test['Na(Peak)'].iloc[0:10]/Rhyo_test['Na(Peak)'].iloc[0], 'sk', mfc='cyan', label='Peak Counts')
ax1.set_xlabel('Time since beam turned on (s)')
ax1.set_ylabel('Peak Counts/Peak Counts in 1st Measurement')

init_I=(Rhyo_test['Na(Peak)'].iloc[0]-0.5*(Rhyo_test['Na(Bg1)'].iloc[0]
        +Rhyo_test['Na(Bg2)'].iloc[0]))
                                          
ax1.plot(Rhyo_test['time since beam turned on'].iloc[0:10], 
         (Rhyo_test['Na(Peak)'].iloc[0:10]-0.5*(Rhyo_test['Na(Bg1)'].iloc[0:10]
        +Rhyo_test['Na(Bg2)'].iloc[0:10]))/init_I, 'ok', mfc='red', label='P- B Counts')
ax1.legend()
fig.savefig('Na_migration_Rhyolite.png', dpi=200)
```

### Tracking declining intensities in Jadiete at 20 nA
- We use exactly the same analytical conditions, but this time measuring the Jadiete primary standard
- While this standard is heterogenous (5-10 wt% Na), as we are tracking intensity on a single point, this is not a concern for this comparison

```python
Jad_test2=pd.read_excel('Supporting_Data_10.xlsx', sheet_name='Jadiete_test2_20nA')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
ax1.set_title('Assesing Na migration in Jadiete at 20 nA')
ax2.set_title('Comparing Na migration in Jadiete and rhyolite')

ax1.plot(Jad_test2['time since beam turned on'].iloc[0:10], 
         Jad_test2['Na(Peak)'].iloc[0:10]/Jad_test2['Na(Peak)'].iloc[0], 'sk', mfc='black', label='Peak Counts - Jadiete')

init_PB=Jad_test2['Na(Peak)'].iloc[0]-0.5*(Jad_test2['Na(Bg1)'].iloc[0]
        +Jad_test2['Na(Bg2)'].iloc[0])
ax1.plot(Jad_test2['time since beam turned on'].iloc[0:10], 
         (Jad_test2['Na(Peak)'].iloc[0:10]-0.5*(Jad_test2['Na(Bg1)'].iloc[0:10]
        +Jad_test2['Na(Bg2)'].iloc[0:10]))/init_PB, 'ok', mfc='red', label='P-B Counts - Jadiete')
ax1.legend()

Reg_Peak=calculate_R2(Jad_test2['time since beam turned on'].iloc[0:10].astype(float), 
         Jad_test2['Na(Peak)'].iloc[0:10].astype(float)/Jad_test2['Na(Peak)'].iloc[0])
ax1.plot(Reg_Peak['x_pred'], Reg_Peak['y_pred'], '-k')

Reg_P_B=calculate_R2(Jad_test2['time since beam turned on'].iloc[0:10].astype(float), 
         (Jad_test2['Na(Peak)'].iloc[0:10].astype(float)
          -0.5*(Jad_test2['Na(Bg1)'].iloc[0:10].astype(float)
        +Jad_test2['Na(Bg2)'].iloc[0:10].astype(float)))/init_PB)
ax1.plot(Reg_P_B['x_pred'], Reg_P_B['y_pred'], '-r')


ax2.plot(Jad_test2['time since beam turned on'].iloc[0:10], 
         Jad_test2['Na(Peak)'].iloc[0:10]/Jad_test2['Na(Peak)'].iloc[0], 'sk',
         mfc='k', label='Peak Counts - Jadiete')

ax2.plot(Rhyo_test['time since beam turned on'].iloc[0:10], 
         Rhyo_test['Na(Peak)'].iloc[0:10]/Rhyo_test['Na(Peak)'].iloc[0], 'ok', 
         mfc='cyan', label='Peak Counts - Rhyolite')


ax2.legend()
ax1.set_ylabel('Counts/Counts in 1st Measurement')
ax1.set_xlabel('Time since beam turned on (s)')
ax2.set_xlabel('Time since beam turned on (s)')

ax1.annotate("R$^{2}$=" +Reg_Peak['R2'], xy=(0.02, 0.7), xycoords="axes fraction", fontsize=12)
ax1.annotate("R$^{2}$=" +Reg_P_B['R2'], xy=(0.02, 0.5), xycoords="axes fraction",
             fontsize=12, color='red')

fig.tight_layout()
ax1.annotate("a)", xy=(0.02, 0.05), xycoords="axes fraction", fontsize=14)
ax2.annotate("b)", xy=(0.02, 0.05), xycoords="axes fraction", fontsize=14)
fig.savefig('Na_migration_Rhyolite_Jadiete_20nA.png', dpi=200)
```

## Method 2 - Repeated analyses of the same point
- A second way to assess migration in Na more quantitatively is to perform repeated analyses at the same stage position. 
- E.g., point 1 cycles through all elements, quantifies them. Then, the acquisition routine starts again, but at the same stage position.
- If elements are migrating under the electron beam, these repeated acquisitions will have different values. 
- For Jadiete, we apply no zero-time-correction, and for Kakanui we do.

## Method 3 - Using the zero-time-intercept tracking function on single points

- This shows that noticable beam migration only happens after ~80s, even inJadiete at 100 nA
