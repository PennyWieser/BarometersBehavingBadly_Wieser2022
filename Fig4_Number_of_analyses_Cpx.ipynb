## Filter for number of analyses

```python
Filt=(cpx_Glass['P_kbar_x']<13) & (cpx_Glass['N_meas_Cpx']>0)
myCpxs1=out['Cpxs'].loc[Filt].reset_index(drop=True)
myLiquids1=out['Liqs'].loc[Filt]

Cpx_Glass=cpx_Glass.loc[Filt].reset_index(drop=True)

# Cpx_Glass['Citation_x']=Cpx_Glass['Citation']
# Cpx_Glass['P_kbar_x']=Cpx_Glass['P_kbar']
# Cpx_Glass['T_K_x']=Cpx_Glass['T_K']
cpx_comps=pt.calculate_clinopyroxene_components(cpx_comps=myCpxs1)
```

## Calculate Cpx-only pressures

```python
Cpx_only_Calcs=pt.calculate_cpx_only_press_all_eqs(cpx_comps=myCpxs1, H2O_Liq=Cpx_Glass['H2O_Liq'])
```

```python
Diff_Wang=Cpx_Glass['P_kbar_x'].reset_index(drop=True)-Cpx_only_Calcs['P_Wang21_eq1'].reset_index(drop=True)
Diff_Put=Cpx_Glass['P_kbar_x'].reset_index(drop=True)-Cpx_only_Calcs['P_Put_Teq32d_Peq32b'].reset_index(drop=True)
```

## Figure for the paper

```python
ms2=7

percent_Wang=np.empty((int(max(Cpx_Glass['N_meas_Cpx']))), float)
percent_Put=np.empty((int(max(Cpx_Glass['N_meas_Cpx']))), float)
Mean_Wang=np.empty((int(max(Cpx_Glass['N_meas_Cpx']))), float)
Mean_Put=np.empty((int(max(Cpx_Glass['N_meas_Cpx']))), float)

N=np.empty((int(max(Cpx_Glass['N_meas_Cpx']))), float)
for i in range(0, int(max(Cpx_Glass['N_meas_Cpx']))):
    if sum(Cpx_Glass['N_meas_Cpx']==i)>4:
        N[i]=i
        
        percent_Wang[i]=np.nanpercentile(np.abs(Diff_Wang.loc[Cpx_Glass['N_meas_Cpx']==i]),  q=95, interpolation='linear')
        percent_Put[i]=np.nanpercentile(np.abs(Diff_Put.loc[Cpx_Glass['N_meas_Cpx']==i]),  q=95, interpolation='linear')
        Mean_Wang[i]=np.nanmean(np.abs(Diff_Wang.loc[Cpx_Glass['N_meas_Cpx']==i]))
        Mean_Put[i]=np.nanmean(np.abs(Diff_Put.loc[Cpx_Glass['N_meas_Cpx']==i]))

        #plt.plot(i,  np.nanmean(np.abs(Diff.loc[Cpx_Glass['N_meas_Cpx']==i])), 'dk', mfc='black', ms=ms2)
    else:
        N[i]=np.nan
        percent_Wang[i]=np.nan
        percent_Put[i]=np.nan
Stats_95_Wang_perc=pt.calculate_R2_np(N, percent_Wang) 
Stats_95_Put_perc=pt.calculate_R2_np(N, percent_Put) 
Stats_95_Wang_Mean=pt.calculate_R2_np(N, Mean_Wang) 
Stats_95_Put_Mean=pt.calculate_R2_np(N, Mean_Put) 


plt.plot(Stats_95_Wang_perc['x_pred'], Stats_95_Wang_perc['y_pred'], '-y')
x=np.linspace(np.nanmax(N), 40, 30)
y_Wang_perc=x*Stats_95_Wang_perc['Grad']+Stats_95_Wang_perc['Int']
y_Put_perc=x*Stats_95_Put_perc['Grad']+Stats_95_Put_perc['Int']
y_Wang_Mean=x*Stats_95_Wang_Mean['Grad']+Stats_95_Wang_Mean['Int']
y_Put_Mean=x*Stats_95_Put_Mean['Grad']+Stats_95_Put_Mean['Int']
plt.plot(x, y_Wang_perc[0], ':y')
print('Wang - 95% perc')
print(Stats_95_Wang_perc['R2'])
print('Put - 95% perc')
print(Stats_95_Put_perc['R2'])
print('Wang - Mean')
print(Stats_95_Wang_Mean['R2'])
print('Put - Mean')
print(Stats_95_Put_Mean['R2'])
```
